```python
import os
import glob
import psycopg2
import pandas as pd
```

```python
# Add path to events log
log_path = 'data/log_data/2018/11/'
```

```python
# Add all log file names to list
all_files = []

for root, dirs, files in os.walk(log_path):
        files = glob.glob(os.path.join(root,'*.json'))
        for f in files :
            all_files.append(os.path.abspath(f))
```

```python
# Extract file contents from jsons

df_log = pd.DataFrame()

for ind in range(len(all_files)):
    df_new_file = pd.read_json(all_files[ind], lines=True)
    df_log = df_log.append(df_new_file, ignore_index=True)

df_log.head(3)
```

### Check whether there are different values in 'level' other than free

```python
df_log[df_log['level'] != 'free'].head(3)
```

### check how many unique values are in level

```python
# print number of unique values in level
print('number of unique values in level: ', df_log.level.nunique())
```
